#### Importing Required modules

```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime
import wikipedia
import pyjokes
import sys
import webbrowser
import os
```

#### Functioning

```python
engine = pyttsx3.init('sapi5')
voice = engine.getProperty('voices')
engine.setProperty('voices', voice[0].id)
engine.setProperty('rate', 150)


# ##################################    TEXT TO SPEECH    ################################## #


def speak(audio):
    engine.say(audio)
    engine.runAndWait()


# ##################################    SPEECH TO TEXT    ################################## #


def takecommand():
    r = sr.Recognizer()
    query = ''
    with sr.Microphone() as source:
        print("Listening...")
        r.pause_threshold = 1
        audio = r.listen(source, timeout=1, phrase_time_limit=5)

    try:
        print('Recognising...')
        query = r.recognize_google(audio, language='en-in')
        if 'Darren' in query:
            query = query.replace('Darren', '')
    except Exception as e:
        # speak('Say that again please.....')
        return "none"
    print('user said:', query)
    return query.lower()


# ##################################    WISH    ################################## #


def wish():
    hour = int(datetime.datetime.now().hour)

    if hour >= 0 and hour < 12:
        print('Good Morning sir!')
        speak('Good morning sir!')
    elif hour >= 12 and hour <= 16:
        print('Good Afternoon sir!')
        speak('Good afternoon sir!')
    elif hour >= 16 and hour <= 20:
        print('Good Evening sir!')
        speak('Good evening sir!')
    else:
        print('Good Night sir!')
        speak('Good night sir!')

    today = datetime.date.today()
    time = datetime.datetime.now().strftime('%I:%M %p')
    # Textual month, day and year
    d2 = today.strftime("%B %d, %Y")
    print("Today's Date is ", d2, 'Current time is', time)
    speak('Today is : ' + d2)
    speak('and current time is ' + time)

    speak('I am Darren, how can I help you?')


# ##################################    TASKS    ################################## #


def taskexecution():
    wish()
    while True:
        command = takecommand()
        if 'hello' in command:
            print('Hello, how can I help you ?')
            speak('hello, how can I help you ?')
        elif 'how are you' in command:
            print('I am fine!\nWhat about you?')
            speak('I am fine, What about you')

        elif 'who are you' in command:
            print('I am Darren a k a your virtual assistant master')
            speak('I am Darren a k a your virtual assistant master. how can i help you ??')

        elif 'sleep now' in command:
            print('Okay sir! You can call me anytime.')
            speak('Ok sir, you can call me anytime')
            break

        elif 'open cmd' in command:
            os.system('start cmd')
        elif 'close cmd' in command:
            os.system('taskkill /f /im cmd.exe')

        elif 'open notepad' in command:
            os.startfile('C:\\ProgramData\\Microsoft\\Windows\\Start Menu\\Programs\\Accessories\\notepad')

        elif 'close notepad' in command:
            os.system('taskkill /f /im notepad.exe')

        elif 'shutdown' in command:
            os.system('shutdown /s /t 5')

        elif 'restart' in command:
            os.system('shutdown /r /t 5')

        elif 'can you do' in command:
            print('''
There are many tasks, I can perform:
     1.  I can play songs on youtube. 
     2.  Tell you a joke.
     3.  Search on wikipedia.
     4.  Tell date and time.
     5.  Find your location.
     6.  Locate area on map.
     7.  Open different websites like Instagram, YouTube, Gmail, github, stack overflow. 
     8.  Search something on google.
     9.  Open and close notepad.
     10. Open and close command prompt.
     11. Shutdown or restart the system.
You can also add some more tasks by doing some modifications in me.
How may I help you ??
             ''')
            speak('''There are many tasks I can perform
            i can play songs on youtube , tell you a joke , search on wikipedia , tell date and time , find your location , locate area on map,
         open different websites like insta gram, youtube, gmail, git hub, stack overflow , search something on google , open and close notepad , open and close command prompt , Shut down or restart the system 
        You can also add some more tasks by doing some modifications in me.
        How may i help you ??''')

        elif 'play' in command:
            song = command.replace('play', '')
            print('Playing' + song)
            speak('Playing' + song)
            pywhatkit.playonyt(song)


        elif 'time' in command:
            time = datetime.datetime.now().strftime('%I:%M %p')
            print('The current time is' + time)
            speak('The current time is')
            speak(time)

        elif 'date' in command:
            today = datetime.date.today()
            print("Today's date:", today)
            # Textual month, day and year
            d2 = today.strftime("%B %d, %Y")
            print("Today's Date is ", d2)
            speak('The todays date is')
            speak(d2)

        elif 'tell me about' in command:
            name = command.replace('tell me about', '')
            info = wikipedia.summary(name, 1)
            print(info)
            speak(info)

        elif 'wikipedia' in command:
            name = command.replace('wikipedia', '')
            info = wikipedia.summary(name, 1)
            print(info)
            speak(info)
        elif 'what is' in command:
            name = command.replace('what is ', '')
            info = wikipedia.summary(name, 1)
            print(info)
            speak(info)

        elif 'who is ' in command:
            name = command.replace('who is', '')
            info = wikipedia.summary(name, 1)
            print(info)
            speak(info)

        elif 'what is ' in command:
            search = 'https://www.google.com/search?q=' + command
            print('searching... Here is what i found on the internet..')
            speak('Here is what i found on the internet..')
            webbrowser.open(search)

        elif 'joke' in command:
            _joke = pyjokes.get_joke()
            print(_joke)
            speak(_joke)

        elif 'search' in command:
            search = 'https://www.google.com/search?q=' + command
            speak('searching... ')
            webbrowser.open(search)

        elif "my location" in command:
            url = "https://www.google.com/maps/search/Where+am+I+?/"
            webbrowser.get().open(url)
            speak("You must be somewhere near here, as per Google maps")
        elif 'locate ' in command:
            speak('locating ...')
            loc = command.replace('locate', '')
            if 'on map' in loc:
                loc = loc.replace('on map', ' ')
            url = 'https://google.nl/maps/place/' + loc + '/&amp;'
            webbrowser.get().open(url)
            print('Here is the location of ' + loc)
            speak('Here is the location of ' + loc)

        elif 'on map' in command:
            speak('locating ...')
            loc = command.split(" ")
            print(loc[1])
            url = 'https://google.nl/maps/place/' + loc[1] + '/&amp;'
            webbrowser.get().open(url)
            print('Here is the location of ' + loc[1])
            speak('Here is the location of ' + loc[1])

        elif 'location of' in command:
            speak('locating ...')
            loc = command.replace('find location of', '')
            url = 'https://google.nl/maps/place/' + loc + '/&amp;'
            webbrowser.get().open(url)
            print('Here is the location of ' + loc)
            speak('Here is the location of ' + loc)

        elif 'where is ' in command:
            speak('locating ...')
            loc = command.replace('where is', '')
            url = 'https://google.nl/maps/place/' + loc + '/&amp;'
            webbrowser.get().open(url)
            print('Here is the location of ' + loc)
            speak('Here is the location of ' + loc)

        elif 'open google' in command:
            print('opening google ...')
            speak('opening google..')
            webbrowser.open_new('https://www.google.co.in/')

        elif 'gmail' in command:
            print('opening gmail ...')
            speak('opening gmail..')
            webbrowser.open_new('https://mail.google.com/')

        elif 'open youtube' in command:
            print('opening you tube ...')
            speak('opening you tube..')
            webbrowser.open_new('https://www.youtube.com/')
        elif 'open instagram' in command:
            print('opening instagram ...')
            speak('opening insta gram...')
            webbrowser.open_new('https://www.instagram.com/')

        elif 'open stack overflow' in command:
            print('opening stackoverflow ...')
            speak('opening stack overflow...')
            webbrowser.open_new('https://stackoverflow.com/')

        elif 'open github' in command:
            print('opening git hub ...')
            speak('opening git hub...')
            webbrowser.open_new('https://github.com/')

        elif 'bye' in command:
            print('good bye, have a nice day !!')
            speak('good bye, have a nice day !!')
            sys.exit()

        elif 'thank you' in command:
            print("your welcome")
            speak('your welcome')

        elif 'stop' in command:
            print('good bye, have a nice day !!')
            speak('good bye, have a nice day !!')
            sys.exit()

        elif 'tata' in command:
            print('good bye, have a nice day !!')
            speak('good bye, have a nice day !!')
            sys.exit()

        elif 'search' in command:
            print(' Here is what i found on the internet..')
            speak('Here is what i found on the internet..')
            search = 'https://www.google.com/search?q=' + command
            webbrowser.open(search)


# ##################################    MAIN    ################################## #


if __name__ == '__main__':
    while True:
        Command = takecommand()
        if 'wake up' in Command:
            taskexecution()
        elif 'bye' in Command:
            print('Thanks for using me!')
            speak('Thanks for using me!')
            sys.exit()
```
